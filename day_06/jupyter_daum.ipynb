```python
from selenium import webdriver

driver = webdriver.Chrome('/Users/bomilee/Documents/GitHub/python_basic/day_06/chromedriver')
driver.get('https://daum.net')
driver.save_screenshot('daumlogin1.png')
```

```python
xpath = '''//*[@id="inner_login"]/a[1]'''
driver.find_element_by_xpath(xpath).click()
driver.save_screenshot('daumlogin2.png')
```

```python
elem_login = driver.find_element_by_id('id')
elem_login.clear()
elem_login.send_keys('id')#아이디입력

elem_login = driver.find_element_by_id('inputPwd')
elem_login.clear()
elem_login.send_keys('pw!')#비밀번호입력
```

```python
xpath = '''//*[@id="loginBtn"]'''
driver.find_element_by_xpath(xpath).click()
driver.save_screenshot('daumlogin3.png')
```
